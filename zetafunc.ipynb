```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = 1
stop = 10000000
s = -10
n = list(range(start,stop))
na = np.asarray(n,dtype=float)

fn = 1/(na**s)

totalna = np.cumsum(fn,axis=0)
#print(fn)
#print(' ')
#print(totalna)

plt.plot(na,totalna)

plt.title('convergent function where s = %g' %(s))
plt.show()
```
